Some descriptive statistics on the raw dataset

Let's do height first. Discriminate between feet (3 digits ft/in,in) and kg (4 digits 9/m/cm,cm)

```python
height = height.tolist() ##convert to a list first
formated_height = []
for elem in height:
    h = str(elem) #convert height to a string
    
    if elem < 1:
        inches = np.nan
    
    if elem > 205.0 and elem < 801.0: ##if standard
        feet = float(h[0])
        inches = float(h[1:])
        if inches > 12: #no one is x ft, 13 in. tall!
            inches = np.nan
        else:
            inches = inches + feet*12.0
            
    elif elem > 9000 and elem < 9999: ##if metric
        meters = float(h[1])
        centim = float(h[2:])
        centim = centim + meters*100
        inches = centim/2.54
        
    else: # if you got this far, you were recorded as a NaN, or were set as a NaN above!
        inches = np.nan
        
    formated_height.append(inches) #Note, everything is in inches!
    
```

Now convert the height back to a Series object

```python
height = pd.Series(formated_height,name='height3')
```

```python
height.value_counts(sort=True, dropna = False) #check thant number of nans = nans from before + 7777s and 9999s
```

Ok, now put the weight in consistent units

```python
weight = weight.tolist()
formated_weight = []
for elem in weight:
    if elem < 1000: #if in standard units
        pounds = elem
        if elem < 10:
            pounds = np.nan
        
    elif elem > 9000 and elem < 9999: #if in metric units
        w = str(elem)
        pounds = float(w[1:])*2.20462 #convert kg to pounds
        if pounds < 10:
            pounds = np.nan
        
    else:    #if you got this far you must have been recorded as a nan or were set as a nan above
        pounds = np.nan 
        
    formated_weight.append(pounds)
        
    
```

Ok, now do the do days per week or per month have at least one drink

```python
alcohol = alcohol.tolist()
formated_alcohol = []
for elem in alcohol:
    dpt = str(elem)
    if elem < 200: #recorded in days per week
        numperweek = float(dpt[1:])
        numpermonth = numperweek*30.0/7.0
        
    elif elem > 200: #recorded in days per month
        numpermonth = float(dpt[1:])
        
    else:
        numpermonth = np.nan
        
    formated_alcohol.append(numpermonth) #final conversion is to numpermonth
        
```
